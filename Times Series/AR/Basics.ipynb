#### Plottng The Data

```python
#One Single Column
df_comp.spx.plot()
plt.title('S&P 500')
plt.show()
```

```python
#One Single Column
df_comp.ftse.plot()
plt.title('FTSE')
plt.show()
```

```python
#One Single Column
df_comp.dax.plot()
plt.title('DAX')
plt.show()
```

```python
#One Single Column
df_comp.nikkei.plot()
plt.title('NIKKEI')
plt.show()
```

```python
#One Single Column
df_comp.spx.plot(figsize=(20,5), title='S&P 500')
plt.show()
```

```python
df_comp.spx.plot(figsize=(20,5), title='S&P 500')
df_comp.ftse.plot(figsize=(20,5), title='FTSE')
plt.title('S&P 500 v FTSE')
plt.show()
```

```python
df_comp.dax.plot(figsize=(20,5), title='DAX')
df_comp.nikkei.plot(figsize=(20,5), title='NIKKEI')
plt.title('DAX v NIKKEI')
plt.show()
```
