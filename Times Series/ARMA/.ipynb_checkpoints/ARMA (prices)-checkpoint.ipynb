```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.tsa.stattools as sts
from statsmodels.tsa.arima_model import ARMA
from scipy.stats.distributions import chi2
import statsmodels.graphics.tsaplots as sgt
from math import sqrt
import seaborn as sns
sns.set()
```

### Data Preprocessing

```python
raw_csv_data = pd.read_csv('013 IndexE81.csv')
df_comp = raw_csv_data.copy()
#Converting Time
df_comp.date = pd.to_datetime(df_comp.date, dayfirst=True)
#Setting Date as an index
df_comp.set_index('date', inplace=True)
#Setting Frequency as business working days
df_comp = df_comp.asfreq('b')
#Filling the Missing or Empty Values
df_comp = df_comp.fillna(method='ffill')
df_comp['market_value'] = df_comp.ftse
```

### Data Splitting

```python
#Splitting Data into train and test
size = int(len(df_comp)*0.8)
df, df_test = df_comp[size:], df_comp[:size]
```

## Autoregressive Moving Average ARMA(P,Q), P = AR, Q = MA

```python
#Model
#ARMA(1,1)
model_ar_1_ma_1 = ARMA(df.market_value, order=(1,1))
results_ar_1_ma_1 = model_ar_1_ma_1.fit()
print(results_ar_1_ma_1.summary())
```

```python
#Filtering Warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
df['results_ar_1_ma_1'] = results_ar_1_ma_1.resid
```

```python
#ADFULLER TEST
sts.adfuller(df.results_ar_1_ma_1)
```

```python
df.results_ar_1_ma_1.plot(figsize=(20,5))
plt.title('Returns', size=24)
plt.show()
```

```python
sgt.plot_acf(df.results_ar_1_ma_1, zero=False, lags=40)
plt.title('ACF for Price Residuals', Size=24)
plt.show()
```
